```python
%matplotlib inline 
import numpy as np 
import matplotlib.pyplot as plt
from ipywidgets import interactive

x= np.linspace(0.1,20,10000)


def LJ(x,eps,sig):
    return (4*eps*(((sig/x)**12)-((sig/x)**6)))
    
def plotting(eps=5,sig=5):
    plt.plot(x,LJ(x,eps,sig))
    plt.ylim(ymax=6,ymin=-6)
    plt.xlim(xmax=12,xmin=0)
    plt.grid()
    plt.xlabel('Distance, r (A)')
    plt.ylabel('Potential, V (kJ/mol)')


interactive_plot = interactive(plotting,eps=(1,5,0.1),sig=(1,5,0.1))
interactive_plot

```
